# Implementasi Python Agglomerative Hierarchical Clustering

### Import Library

```python
# Import Library
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np

# Buat Array 2D ke dalam variabel x. Data ini yang akan kita Klaster
x = np.array([[1,2], [2.5,4.5], [2,2], [4,1.5], [4,2.5]])

# Lakukan Hierarchical Klastering
Z = linkage(x)

# Plot Dendogram
plt.figure(figsize=(25, 10))
plt.title('Dendogram Hierarchical Klastering')
plt.xlabel('data')
plt.ylabel('jarak')
dendrogram(Z)
plt.show()
```
